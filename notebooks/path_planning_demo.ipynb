```python
# Cell 1: Path Planning Visualization in Jupyter
# Run: jupyter notebook notebooks/path_planning_demo.ipynb

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import sys
sys.path.append('..')  # Allow imports from project root

from src.path_planner import PathPlanner

# ── Create a test grid ────────────────────────────────────────
ROWS, COLS = 15, 20
grid = np.zeros((ROWS, COLS), dtype=int)

# Add border
grid[0, :]    = 1
grid[-1, :]   = 1
grid[:, 0]    = 1
grid[:, -1]   = 1

# Add some obstacles
obstacles = [
    (5,2),(5,3),(5,4),(5,5),(5,6),
    (10,8),(10,9),(10,10),(10,11),(10,12),
    (3,10),(4,10),(5,10),(6,10),
    (15,3),(15,4),(15,5),(16,3),(16,4)
]
for (c,r) in obstacles:
    if 0 < r < ROWS-1 and 0 < c < COLS-1:
        grid[r][c] = 1

start = (1, 1)
goal  = (18, 13)

# ── Run A* ────────────────────────────────────────────────────
planner = PathPlanner()
path = planner.find_path(start, goal, grid)
print(f"Path found: {len(path)} waypoints")
print(f"Nodes explored: {planner.nodes_explored}")

# ── Visualize ─────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

for ax, title, show_path in [(axes[0], "Grid + A* Path", True),
                              (axes[1], "Exploration Visualization", True)]:
    ax.set_xlim(0, COLS)
    ax.set_ylim(0, ROWS)
    ax.set_aspect('equal')
    ax.set_title(title, fontsize=14, fontweight='bold')
    ax.set_facecolor('#1a1a2e')

    # Draw grid cells
    for r in range(ROWS):
        for c in range(COLS):
            color = '#4a4a4a' if grid[r][c] == 1 else '#16213e'
            rect = patches.Rectangle((c, ROWS-1-r), 1, 1,
                                      linewidth=0.3, edgecolor='#2a2a3e',
                                      facecolor=color)
            ax.add_patch(rect)

    # Draw path
    if path and show_path:
        px = [c + 0.5 for (c,r) in path]
        py = [ROWS - r - 0.5 for (c,r) in path]
        ax.plot(px, py, 'o-', color='#0096ff', linewidth=2,
                markersize=4, label=f'A* Path ({len(path)} steps)', zorder=3)

    # Start and Goal
    sc, sr = start
    gc, gr = goal
    ax.scatter([sc+0.5], [ROWS-sr-0.5], c='#00ff88', s=200,
               zorder=5, label='Start', marker='*')
    ax.scatter([gc+0.5], [ROWS-gr-0.5], c='#ff4444', s=200,
               zorder=5, label='Goal', marker='X')

    ax.legend(loc='upper right', fontsize=9)
    ax.set_xticks(range(COLS))
    ax.set_yticks(range(ROWS))
    ax.tick_params(labelsize=7)

plt.suptitle("A* Pathfinding Algorithm — AI Autonomous Navigation",
             fontsize=16, fontweight='bold', color='white',
             y=1.01)
plt.tight_layout()
plt.savefig('../images/path_planning_demo.png', dpi=150,
            bbox_inches='tight', facecolor='#0d0d1a')
plt.show()

print(f"\nPath waypoints: {path}")
```
